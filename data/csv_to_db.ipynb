```python
def add_parentheses(value):
    return f'({value}),' #f"('{value}'),"  #f'("{value}"),'                 

# apply the function to each value in the product_id column
data["product_id"]=data["product_id"].apply(add_parentheses)
data["product_category"]=data["product_category"].apply(add_parentheses)
data["Height"]=data["Height"].apply(add_parentheses)
data["width"]=data["width"].apply(add_parentheses)
data["material"]=data["material"].apply(add_parentheses)
data["color"]=data["color"].apply(add_parentheses)
data["product_image"]=data["product_image"].apply(add_parentheses)
data["product_brand"]=data["product_brand"].apply(add_parentheses)
data["product_price"]=data["product_price"].apply(add_parentheses)
data["product_offer"]=data["product_offer"].apply(add_parentheses)
data["product_rate"]=data["product_rate"].apply(add_parentheses)
data["seller_id"]=data["seller_id"].apply(add_parentheses)
data["product_name"]=data["product_name"].apply(add_parentheses)
data["product_description"]=data["product_description"].apply(add_parentheses)
data["depth"]=data["depth"].apply(add_parentheses)
```
